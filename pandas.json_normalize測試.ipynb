```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
api_url = 'https://opendata.cwa.gov.tw/api/v1/rest/datastore/F-C0032-001'
api_params = {
    'Authorization' : CWA_KEY,
}
response = requests.get(api_url, params = api_params)
# Check if the API request was successful
response.raise_for_status()
```

```python
response.json()
```

```python
input_data = response.json()
```

```python
input_data['records']
```

```python
from pandas import json_normalize
```

```python
# record_path指定想要展開的主要資料路徑
normalized_df = json_normalize(data=input_data['records'],
                              record_path=['location', 'weatherElement'])
```

```python
normalized_df
```

```python
normalized_df = json_normalize(data=input_data['records'],
                              record_path=['location', 'weatherElement', 'time'])
```

```python
normalized_df
```

```python
# meta允許從外部級別提取其他資訊
normalized_df = json_normalize(data=input_data['records'],
                              record_path=['location', 'weatherElement', 'time'],
                              meta=[['location', 'locationName'], ['location','weatherElement', 'elementName']])
```

```python
normalized_df
```

```python
# record_prefix與mata_prefix為展開的資料與元資料添加前綴，使欄位名稱更易說明
normalized_df = json_normalize(data=input_data['records'],
                              record_path=['location', 'weatherElement', 'time'],
                              meta=[['location', 'locationName'], ['location','weatherElement', 'elementName']],
                              record_prefix='time_',
                              meta_prefix='meta_')
```

```python
normalized_df
```
